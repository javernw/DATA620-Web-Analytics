### In this assignment, create the nodes and edges for a basic graph, such as the Krackhardt kite.

```python
#import packages

import networkx as nx
import matplotlib.pyplot as plt
% matplotlib inline

# create a dictionary with the details for the graph

myDict = {
    "Carol": ["Andre", "Diane", "Fernando"],
    "Andre": ["Beverley", "Diane", "Fernando", "Carol"],
    "Fernando": ["Heather", "Garth", "Carol", "Diane", "Andre"],
    "Diane": ["Andre", "Carol", "Beverley", "Fernando", "Ed", "Garth"],
    "Beverley": ["Ed", "Garth", "Diane", "Andre"],
    "Ed": ["Garth", "Beverley", "Diane"],
    "Garth": ["Fernando", "Diane", "Beverley", "Ed"],
    "Heather": ["Fernando", "Garth", "Ike"],
    "Ike": ["Heather", "Jane"],
    "Jane": ["Ike"]
}

# create graph from dictionary
myGraph = nx.Graph(myDict)

# show graph information
print nx.info(myGraph)

# draw graph with the labels
nx.draw(myGraph, with_labels=True)

# display graph
plt.show(myGraph)



```
